```python
import numpy as np
import pandas as pd
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
data=pd.read_csv('/content/movies.csv')
```

```python
print(data.head())
```

```python
 data.shape
```

```python
selected_features=['genres','keywords','tagline','cast','director']
```

```python
for feature in selected_features:
  data[feature]=data[feature].fillna('')
  
```

```python
new_feature=data['genres']+' '+data['keywords']+' '+data['tagline']+' '+data['cast']+' '+data['director']
```

```python
print(new_feature.head())
```

```python
vectorizer=TfidfVectorizer()
feature_vector=vectorizer.fit_transform(new_feature)
```

```python
print(feature_vector)
```

```python
#similarity score 
similarity=cosine_similarity(feature_vector)
```

```python
print(similarity)
```

```python
print(similarity.shape)
```

```python
movie_name=input("enter your favourite movie name :")
```

```python
#getting all movie names  in a list 
movie_list=data['title'].tolist()
print(movie_list)
```

```python
#finding close match 
close_match=difflib.get_close_matches(movie_name,movie_list)
print(close_match)
```

```python
final_match=close_match[0]
print(final_match)
```

```python
#movie index
movie_index=data[data.title==final_match]['index'].values[0]
print(movie_index)
```

```python
similarity_score=list(enumerate(similarity[movie_index]))
print(similarity_score)
```

```python
len(similarity_score)
```

```python
similar_movies=sorted(similarity_score,key=lambda x:x[1], reverse=True)
```

```python
print(similar_movies)
```

```python
print('movies suggested for you : \n')
i=1
for movie in similar_movies:
  index=movie[0]
  title_from_index=data[data.index==index]['title'].values[0]
  if(i<30):
    print(i,'.',title_from_index)
    i+=1
    
```
